01 Matplotlib 기본 사용

```python
import matplotlib.pyplot as plt
```

```python
plt.plot([1,2,3,4])
plt.show()
```

```python
plt.plot([1,2,3,4],[1,4,9,16])
plt.show()
```

스타일지정하기

```python
plt.plot([1,2,3,4],[1,4,9,16],'ro')
plt.axis([0,6,0,20])  # [xmin, xmax, ymin, ymax]
plt.show()
```

여러 개의 그래프 그리기

```python
import numpy as np

#0.2 간격으로 균일하게 샘플링
t = np.arange(0.,5.,0.2)

# 빨간 대쉬, 파란 삼각형, 녹색 삼각형
plt.plot(t,t,'r--', t,t**2,'bs', t,t**3,'g^')
plt.show()
```

02. Matplotlib 숫자 입력하기
기본적으로 리스트 하나 넣으면 y 좌표로 인식
넘파이 배열이나 튜플도 입력 가능

```python
plt.plot([2,3,5,10])
plt.show()
```

```python
# x,y 값 입력하기
plt.plot([1,2,3,4], [2,3,5,10])
plt.show()
```

```python
# 레이블이 있는 데이터 사용하기

data_dict = {'data_x':[1,2,3,4,5], 'data_y':[2,3,5,10,8]}

plt.plot('data_x','data_y', data = data_dict)
plt.show()
```

03. Matplotlib 축 레이블 설정하기

keyword: plt.xlabel(), plt.ylabel, plt.axis(), labelpad, fontdict, loc

```python
plt.plot([1,2,3,4],[2,3,5,10])
plt.xlabel('X-Axis' )
plt.ylabel('Y-Axis')
plt.show()
```

```python
plt.plot([1,2,3,4], [2,3,5,10])
plt.xlabel('X-Axis', labelpad = 15)
plt.ylabel('Y-Axis, ', labelpad= 20)
plt.show()
```

```python
# 위치 지정하기, loc=
# xlabel은 left, center, right
# ylabel은 bottom, center, top

plt.plot([1,2,3,4],[2,3,5,10])
plt.xlabel('X-Axis', loc = 'right')
plt.ylabel('Y-Axis', loc = 'top')
```
